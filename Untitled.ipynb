```python
from flask import Flask, render_template, request, jsonify, session
import pandas as pd
import mysql.connector
import json
from dash import Dash
from dash.dependencies import Input, Output
from dash import dcc
from dash import html
import plotly.graph_objs as go
import random
from bs4 import BeautifulSoup

app = Flask(__name__)
# dash_app = Dash(__name__, server=app, url_base_pathname='/final/')

# 
file = open(
    '/Users/seira/Downloads///.txt', "r"
)
js = file.read()
all_stock_id = json.loads(js)
file.close()

# 
cnx = None
cnx2 = None
cnx3 = None

# 
def generate_random_colors(num_colors):
    colors = []
    for _ in range(num_colors):
        r = random.randint(0, 255)
        g = random.randint(0, 255)
        b = random.randint(0, 255)
        a = random.uniform(0.3, 1.0)  # 
        rgba_color = f'rgba({r}, {g}, {b}, {a})'
        colors.append(rgba_color)
    return colors

# MYSQL data()
def get_database_connection():
    global cnx
    if cnx is None:
        cnx = mysql.connector.connect(
            host="127.0.0.1",  # 
            user="root",  # 
            password="seira860816",  # 
            database="data",  # 
        )
    return cnx

# MYSQL data2()
def get_database_connection2():
    global cnx2
    if cnx2 is None:
        cnx2 = mysql.connector.connect(
            host="127.0.0.1",  # 
            user="root",  # 
            password="seira860816",  # 
            database="data2",  # 
        )
    return cnx2

# MYSQL data3(beta)
def get_database_connection3():
    global cnx3
    if cnx3 is None:
        cnx3 = mysql.connector.connect(
            host="127.0.0.1",  # 
            user="root",  # 
            password="seira860816",  # 
            database="data3",  # 
        )
    return cnx3

# beta
def get_stock_beta(stockid):
    query = f'select CAST(`beta` AS DECIMAL(10, 5))as Beta from `{stockid}`'
    cnx3 = get_database_connection3()
    df = pd.read_sql(query, cnx3)
    first_row = df.iloc[0]
    beta_value = first_row["Beta"]
    return beta_value

# 
def get_stock_price(stockid):
    query = f"""select CAST(`open` AS DECIMAL(10, 2)) as price from `{stockid}`
        WHERE `date`  = (
            SELECT MAX(`date`)
            FROM `{stockid}`
        )"""
    cnx = get_database_connection()
    df = pd.read_sql(query, cnx)
    first_row = df.iloc[0]
    price_value = first_row["price"]
    return price_value

# 
def get_stock_name(stockid):
    stock_df = pd.read_csv(
        '/Users/seira/Downloads///.csv')
    matched_stock = stock_df.loc[stock_df[''] == stockid, '']
    if not matched_stock.empty:
        return matched_stock.values[0]
    else:
        return ''

# 
@app.route("/")
def index():
    return render_template("index.html")

# 
@app.route("/stock/<stock>", methods=["GET", "POST"])
def stock(stock):
    if request.method == "GET":
        # 
        stock_code = stock

        # 
        # 
        query = f"""
                SELECT `{stock_code}`.`date` AS , CAST(`Open` AS DECIMAL(10, 2))as , CAST(`High` AS DECIMAL(10, 2))as , CAST(`Low` AS DECIMAL(10, 2))as , CAST(`Close` AS DECIMAL(10, 2))as , CAST(`K9` AS DECIMAL(10, 2))as K, CAST(`D9` AS DECIMAL(10, 2))as D
                FROM `{stock_code}`
                order by `date` DESC
                """

        query2 = f"""
                SELECT  `{stock_code}`.`Unnamed: 0` AS ,
                    CAST(`` AS DECIMAL(10, 2))as ,
                    CAST(`(%)` AS DECIMAL(10, 2))as ,
                    CAST(`(/)` AS DECIMAL(10, 2))as ,
                    CAST(`(/)` AS DECIMAL(10, 2))as ,
                    CAST(`  ()` AS DECIMAL(10, 2))as ,
                    CAST(`  ()` AS DECIMAL(10, 2))as ,
                    CAST(`()` AS DECIMAL(10, 2))as ,
                    CAST(`` AS DECIMAL(10, 2))as ,
                    CAST(`(/)` AS DECIMAL(10, 2))as 

                FROM `{stock_code}`
                """
        stID=stock_code
        stName=get_stock_name(int(stock_code))
        cnx = get_database_connection()
        cnx2 = get_database_connection2()
        df = pd.read_sql(query, cnx)
        df2 = pd.read_sql(query2, cnx2)
        table_html = df.to_html(index=False)
        table2_html = df2.to_html(index=False)
        
        for i in all_stock_id.keys():
            for j in all_stock_id[i]:
                if stock_code==j:
                    file_path = r"/Users/seira/Downloads////" + j + ".html"
                    
                    with open(file_path, 'r', encoding='utf-8') as file:
                        html_content = file.read()
                       



        return render_template(
            "stock.html", stock_code=stock_code, table_html=table_html, table2_html=table2_html,stID=stID,stName=stName
        ,html_content=html_content)
    else:
        return render_template("stock.html")

# 
@app.route("/search", methods=["GET", "POST"])
def result():
    if request.method == "POST":
        # 
        stock_code = request.form["stock"]

        # 
        # 
        try:
            query = f"""
                SELECT `{stock_code}`.`date` as , CAST(`Open` AS DECIMAL(10, 2))as , CAST(`High` AS DECIMAL(10, 2))as , CAST(`Low` AS DECIMAL(10, 2))as , CAST(`Close` AS DECIMAL(10, 2))as , CAST(`K9` AS DECIMAL(10, 2))as K, CAST(`D9` AS DECIMAL(10, 2))as D
                FROM `{stock_code}`
                order by `date` DESC
                """
            # 
            query2 = f"""
                SELECT  `{stock_code}`.`Unnamed: 0` AS ,
                    CAST(`` AS DECIMAL(10, 2))as ,
                    CAST(`(%)` AS DECIMAL(10, 2))as ,
                    CAST(`(/)` AS DECIMAL(10, 2))as ,
                    CAST(`(/)` AS DECIMAL(10, 2))as ,
                    CAST(`  ()` AS DECIMAL(10, 2))as ,
                    CAST(`  ()` AS DECIMAL(10, 2))as ,
                    CAST(`()` AS DECIMAL(10, 2))as ,
                    CAST(`` AS DECIMAL(10, 2))as ,
                    CAST(`(/)` AS DECIMAL(10, 2))as 

                FROM `{stock_code}`
                """

            stID=stock_code
            stName=get_stock_name(int(stock_code))
            cnx = get_database_connection()
            cnx2 = get_database_connection2()
            df = pd.read_sql(query, cnx)
            df2 = pd.read_sql(query2, cnx2)
            table_html = df.to_html(index=False)
            table2_html = df2.to_html(index=False)

            for i in all_stock_id.keys():
                for j in all_stock_id[i]:
                    if stock_code==j:
                        file_path = r"/Users/seira/Downloads////" + j + ".html"
                        
                        with open(file_path, 'r', encoding='utf-8') as file:
                            html_content = file.read()
              

            #  HTML 
            return render_template(
            "stock.html", stock_code=stock_code, table_html=table_html, table2_html=table2_html,stID=stID,stName=stName,html_content=html_content
        )
        except Exception as e:
            print(e)
            error_message = f""
            return render_template("index.html", error_message=error_message)
    else:
        return render_template("index.html")

# STEP1
@app.route("/", methods=["GET", "POST"])
def receive_data():
    global selected_types
    if request.method == "POST":
        selected_types = request.form.getlist("types[]")
        table_html = ""
        rank_value = request.form['rank']
        cnx = get_database_connection()
        cnx2 = get_database_connection2()
        df_combined = pd.DataFrame()  # 
        q = 0  # 

        # (table)
        for name in selected_types:
            for stock_code in all_stock_id[name]:
                q += 1
                name1 = name
                query = f"""
                    SELECT '{stock_code}' AS , '{get_stock_name(int(stock_code))}' AS ,'{name}' AS ``, `{stock_code}`.`date`as , CAST(`Open` AS DECIMAL(10, 2))as , CAST(`High` AS DECIMAL(10, 2))as , CAST(`Low` AS DECIMAL(10, 2))as , CAST(`Close` AS DECIMAL(10, 2))as , CAST(`K9` AS DECIMAL(10, 2))as K, CAST(`D9` AS DECIMAL(10, 2))as D
                    FROM `{stock_code}`
                    WHERE date = (
                        SELECT MAX(date)
                        FROM `{stock_code}`
                    );
                    """
                query2 = f"""
                    SELECT '{stock_code}' AS , `{stock_code}`.`Unnamed: 0` AS , CAST(`{rank_value}` AS DECIMAL(10, 2)) AS `{rank_value}`
                    FROM `{stock_code}`
                    WHERE `Unnamed: 0` = (
                        SELECT MAX(`Unnamed: 0`)
                        FROM `{stock_code}`
                    );
                """

                # 
                df = pd.read_sql(query, cnx)
                df2 = pd.read_sql(query2, cnx2)
                merged_df = pd.merge(df, df2, on='', how='inner')
                df_combined = pd.concat(
                    [df_combined, merged_df], ignore_index=True)  # 
        df_combined[""] = df_combined[""].apply(
            lambda x: f"<input type=checkbox name='search' value='{x}'><a href='/stock/{x}'>{x}</a>")
        # 
        df_sorted = df_combined.sort_values(by=rank_value, ascending=False)
        stock_avg = round(df_sorted[rank_value].mean(), 4)

        # HTML
        if not df_sorted.empty:
            table_html += df_sorted.to_html(header=True,
                                            index=False, escape=False)

        #  HTML 
        return render_template(
            "index.html", stock_code=stock_code, table_html=table_html, q=q, name1=name1, stock_avg=stock_avg, rank_value=rank_value)

    else:
        return render_template("index.html")

# ()
@app.route("/final", methods=["POST"])
def final():
    stockcodes = request.form.getlist("search")  # 
    betavalues = []
    stocknames = []
    stockprices = []
    tablekd_html = ""
    cnx = get_database_connection()
    df_combined = pd.DataFrame()  # 

    # 
    for stockid in stockcodes:
        betavalues.append(get_stock_beta(int(stockid)))
        stocknames.append(get_stock_name(int(stockid)))
        stockprices.append(get_stock_price(int(stockid)))
    # KD
    for code in stockcodes:
        # 
        query = f"""SELECT '{code}' AS ID, '{get_stock_name(int(code))}' AS name, MAX(`date`) as `` FROM `{code}`WHERE `` <> 0;"""
        # 
        query2 = f"""SELECT '{code}' AS ID, MAX(`date`) as `` FROM `{code}`WHERE `` <> 0;"""
        # KD
        query3 = f"""SELECT '{code}' AS ID, `KD1`as `KD`,`KD2`as `` FROM `{code}` WHERE date = (SELECT MAX(date) FROM `{code}`);"""
        df = pd.read_sql(query, cnx)
        df2 = pd.read_sql(query2, cnx)
        df3 = pd.read_sql(query3, cnx)
        merged_df = pd.merge(df, df2, on='ID', how='inner')
        merged_df = pd.merge(merged_df, df3, on='ID', how='inner')
        df_combined = pd.concat(
            [df_combined, merged_df], ignore_index=True)  # 
    mapping1 = {3: '', 1: '', 2: ''}  # 
    mapping2 = {1: 'KD>80 ', 2: 'KD<20 ', 3: "20KD80"}  # KD
    df_combined['KD'] = df_combined['KD'].replace(mapping2)
    df_combined[''] = df_combined[''].replace(mapping1)
    # HTML
    if not df_combined.empty:
        tablekd_html += df_combined.to_html(header=True,
                                            index=False, escape=False)

    stocks_with_beta = zip(stockcodes, betavalues, stocknames, stockprices)

    return render_template("final.html", stocks_with_beta=stocks_with_beta, tablekd_html=tablekd_html)

#
@app.route("/home", methods=["GET", "POST"])
def home():
    return render_template("index.html")

if __name__ == "__main__":
    app.run(port=5555)
    # threaded=True,
```
